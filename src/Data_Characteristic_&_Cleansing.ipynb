```python
#Make columns names in proper format by assigning proper variables
dataset.columns=['Product_ID', 'Category', 'SubCategory1', 'SubCategory2', 'Location',
       'Channel', 'Customer_Age', 'Review_Title', 'Review_Text', 'Rating',
       'Recommend_Flag']
```

```python
# Find duplicate rows
duplicated_rows = dataset.duplicated()

# Count the number of duplicate rows
number_of_duplicate_rows = duplicated_rows.sum()

print("Number of duplicated rows:", number_of_duplicate_rows)
```

```python
#Remove duplicate data
dataset=dataset.drop_duplicates()
```

```python
# Validate with duplicate rows again
duplicated_rows = dataset.duplicated()

# Count the number of duplicate rows
number_of_duplicate_rows = duplicated_rows.sum()

print("Number of duplicated rows:", number_of_duplicate_rows)
```

```python
#Check null values in dataset
dataset.isnull().sum()
```

```python
rows_missing_category = dataset[dataset['Category'].isnull()]
# Display rows with missing 'Category' as a DataFrame
rows_missing_category
```

```python
dataset['Category'].fillna(dataset['Category'].mode()[0], inplace=True)
dataset['SubCategory1'].fillna(dataset['SubCategory1'].mode()[0], inplace=True)
dataset['SubCategory2'].fillna(dataset['SubCategory2'].mode()[0], inplace=True)
```

```python
rows_missing_Review_Text  = dataset[dataset['Review_Text'].isnull()]
# Display rows with missing 'Review_Text ' as a DataFrame
rows_missing_Review_Text 
```

```python
# remove rows where Review_Title and Review_Text have NaN values
dataset.dropna(subset=['Review_Title', 'Review_Text'], how='all', inplace=True)
```

```python
dataset.isnull().sum()
```

```python
rows_missing_Review_Text  = dataset[dataset['Review_Text'].isnull()]
# Display rows with missing 'Review_Text ' as a DataFrame
rows_missing_Review_Text 
```

```python
rows_missing_Review_Title  = dataset[dataset['Review_Title'].isnull()]
# Display rows with missing 'Review_Text ' as a DataFrame
rows_missing_Review_Title
```

```python
import pandas as pd

# Replace "NaN" values in both columns with empty strings
dataset['Review_Text'].fillna('', inplace=True)
dataset['Review_Title'].fillna('', inplace=True)

# Merge the columns and store the result in a new column called "Merged_Review"
dataset['Merged_Review'] = dataset['Review_Title'] + ' ' + dataset['Review_Text']

# Drop the individual "Review_Text" and "Review_Title" columns if needed
dataset.drop(['Review_Text', 'Review_Title'], axis=1, inplace=True)
```
